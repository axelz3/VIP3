```python
import numpy as np
import ps_utils
import numpy.linalg as la
import matplotlib.pyplot as plt

def ex_2_and_3(filename):

    # read Beethoven data
    I, mask, S = ps_utils.read_data_file(filename)

    # get indices of non zero pixels in mask
    nz = np.where(mask > 0)
    m,n = mask.shape

    # for each mask pixel, collect image data
    J = np.zeros((3, len(nz[0])))
    for i in range(3):
        Ii = I[:,:,i]
        J[i,:] = Ii[nz]


    # solve for M = rho*N
    iS = la.inv(S)
    M = np.dot(iS, J)

    # get albedo as norm of M and normalize M
    Rho = la.norm(M, axis=0)
    N = M/np.tile(Rho, (3,1))

    # display the albedo as a 2D image
    albedo = np.zeros((m,n))
    albedo[nz] = Rho
    plt.title("Albedo")
    plt.imshow(albedo)
    plt.show()

    n1 = np.zeros((m,n))
    n2 = np.zeros((m,n))
    n3 = np.ones((m,n))
    n1[nz] = N[0,:]
    n2[nz] = N[1,:]
    n3[nz] = N[2,:]

    _,(ax1,ax2,ax3) = plt.subplots(1,3)
    ax1.imshow(n1)
    ax2.imshow(n2)
    ax3.imshow(n3)
    plt.show()

    z = ps_utils.unbiased_integrate(n1, n2, n3, mask)
    z = np.nan_to_num(z)

    ps_utils.display_surface_matplotlib(z)


ex_2_and_3('Beethoven') # exercise 2
ex_2_and_3('mat_vase') # exercise 3

```
